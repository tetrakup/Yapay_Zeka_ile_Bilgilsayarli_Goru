```python
import cv2
import numpy as np
```

```python
with open("data/coco.txt", "r") as l:
    class_names = l.read().split("\n")
class_names
```

```python
colors = np.random.uniform(0, 255, size = (len(class_names), 3))
```

```python
model = cv2.dnn.readNet(
    model = "data/frozen_inference_graph.pb",
    config = "data/ssd_mobilenet_v2_coco_2018_03_29.pbtxt.txt",
    framework = "Tensorflow",
)
```

```python
cap  = cv2.VideoCapture("test.mp4") #webcam --> cv2.VideoCapture(0)
#height, width, chanell = img.shape

while True:
    ret, frame = cap.read()

    if ret == False:
        break
    height, width, chanell = frame.shape

    blob = cv2.dnn.blobFromImage(
                                 image = frame,
                                 size = (300,300),
                                 mean = (104, 117, 123),
                                 swapRB = True
                                 )
    
    model.setInput(blob)
    output = model.forward()

    for detection in output[0,0, :, :]:

        confidence = detection[2]

        if confidence > .5:
            class_id = detection[1]
            class_name = class_names[int(class_id)-1]
            color = colors[int(class_id)]

            x = int(detection[3]* width)
            y = int(detection[4] * height)

            x_width = int(detection[5]* width)
            y_height = int(detection[6] * height)

            cv2.rectangle(frame, (x,y), (x_width, y_height), color, thickness=2)
            cv2.putText(frame, class_name, (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
            
    cv2.imshow("Prediction", frame)
    
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows


```
