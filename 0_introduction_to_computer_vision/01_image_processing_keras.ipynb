```python
#library #tensorflow

#from keras.preprocessing.image import load_image
from tensorflow.keras.utils import load_img
```

```python
#reading image/loading --> load_image
img = load_img("datasets/leon.jpg")

#type
print(img, end="\n\n")
print("Type:",type(img))
print("Format:",img.format)
print("Size:",img.size)
print("Mode:",img.mode)
```

```python
#displaying image
%pip install matplotlib
import matplotlib.pyplot as plt
img = load_img("datasets/leon.jpg")

plt.imshow(img)
plt.show
```

```python
#PIL --> numpy array
from tensorflow.keras.utils import load_img, img_to_array, array_to_img

img = load_img("datasets/leon.jpg")
print("Type:",type(img))

img = img_to_array(img)
print("Type:",type(img))

# numpy array --> PIL
img = array_to_img(img)
print("Type:",type(img))
```
